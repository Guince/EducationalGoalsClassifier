В столбцах представлена такая информация:

__label_id__ [1, 18038]: id уникальной разметки (которых ~18 тысяч)

__label_form_id__ [0, 5999]: фиг знает

__label_user_id__ [0, 14]: id разметчика. К user_id привязана РОЛЬ разметчика. Поэтому у одного человека было несколько user_id, отвечающих разным ролям.

__label_goal_id__ [0, 5999]: ТОТ САМЫЙ уникальный идентификатор цели.

__label_num__ [0, 1]:  вообще фиг знает. Видимо бесполезный столбец, так как label_num равен 0 только в 2 строках.

***

У всех следующих признаков область значений такая: [0, 1]

Но встречаются и значения -1. -1 значит

__label_specific__ : более детальное уточнение по самой цели, хоть что-то связанное с предметной областью, к которой цель имеет отношение (конкретность цели)

__label_attainable__ : достижимость + данные из контекста по первым шагам и преградам. Достижимость в принципе для данного человека

__label_time_bound__ : здесь нужны фильтры на наличие категорий времени (слова-маркеры) + РАЗДЕЛИТЬ ограниченность по сроку и по количеству часов (знаю до когда сделать, не знаю, сколько буду делать) ‘ это в анкету

__label_education__ : если понятно, чему учить, то да. Непонятно – то нет. Этот пункт нужен, чтобы отсеивать не-образовательные цели, цель относится к образованию (не слава, здоровье и проч.) - в качестве дополнительного критерия

__label_unambiguity__ : не разделяется на несколько самостоятельных целей (но при этом разделяется на этапы)

***

У всех следующих признаков область значений такая: [Да, Нет, Не следует из формулировки, но можно предположить] 

Каждой цели назначается не менее двух, и не более четырех аттракторов. Важно отмечать не менее двух аттракторов, чтобы не заужать пользователя на одну область, а иметь возможность расширить спектр рекомендаций. 

__label_attractor_knowledge__ : Получение новых знаний в любой области.

Примеры: "Разобраться в основных математических моделях", "Изучить историю России", "Научится разбираться в футбольных командах"

__label_attractor_hard_skill__ : Навык, для реализации которого нужен значительный объем знаний

__label_attractor_soft_skill__ : Навык, для которого часто нужно проживать какие-то состояния, включаться в среду, проделывать универсальные когнитивные шаги

__label_attractor_tool__ : Освоить новые инструменты деятельности

__label_attractor_community__ : Вступить в какое-либо сообщество, стать его частью


__label_attractor_subjectivity__ : Всё, что касается субъектности, с самоосознанием и самоощущением, изменением самости человека

__label_attractor_habits__ : Регулярно повторяющиеся действия, приносящие пользу, или устранение негативных привычек

__label_attractor_career__ : Всё, что касается трудоустройства, карьерного продвижения, перехода к новой должности

__label_attractor_fixing__ : Жизненные достижения/стадии, которые закрепляются формальными документами

__label_attractor_art__ : Всё, что касается или творчества в привычном понимании, или создания чего-то нового

__label_attractor_health__ : Описания нет, но очевидно всё, что касается здоровья и ЗОЖ

***

__label_abstraction_level__ [Конкретная цель, Цель в предметной области, Абстрактная цель] : Пользователь указывает не просто цели в общем, а цели самообразования и саморазвития, поэтому каждая цель имеет уровень абстракции относительно образовательного результата. Таким результатом выступает конкретное знание (знаниевый элемент предметной области) или навык (предметный, инструментальный или soft-skill). Каждой цели приписывается только один уровень абстракции из трех.

Уровни абстракции:

- Конкретная цель:
Цель пользователя можно отнести к конкретному полю тематического рубрикатора (автоматически попадает на таксономию*)

Пример: "Получить водительские права", "Повысить навык публичных выступлений и выступлений перед камерой",
"Изучить С++"

- Цель в предметной области:
Пользователь не знает, чего именно хочет достичь, но представляет и указывает предметную область

Пример:
"Разобраться в основных математических моделях", "Изучить историю России", "Освоить новые программные продукты"

- Абстрактная цель:
Пользователь указывает свою цель в очень общих категориях, на отрицание/избегание; цель = абстрактная хотелка

Пример: "Быть автономной, понимать объективную истину (насколько это возможно) и не зависеть от мнения окружающих", "Быть счастливой", "Самосовершенствование", "Не останавливаться в развитии2


__label_questions__ : Вопросы, которые бы мог задать чат бот для уточнения цели. Эти вопросы можно было выбрать из списка предложенных.

__label_custom_question__ : Вопросы, которые бы мог задать чат бот для уточнения цели. Этот вопрос разметчик должен был написать сам.

__label_with_context__ [0, 1]:  0 - цель размечалась без знания об ответах на вопросы, только голая цель. 1 - разметчик видел ответы респондента на вопросы

Убираем информацию из анкеты для таких целей, которые разметчики размечали не видя информации из анкеты.

```python
for i in info.index:
    if (dataset.loc[i, 'label_with_context'] == 0):
        info.loc[i, 'goal_type_tokens'] = 'НЕТ ИНФОРМАЦИИ'
        info.loc[i, 'first_step_tokens'] = 'НЕТ ИНФОРМАЦИИ'
        info.loc[i, 'time_tokens'] = 'НЕТ ИНФОРМАЦИИ'
        info.loc[i, 'goal_field_tokens'] = 'НЕТ ИНФОРМАЦИИ'
        info.loc[i, 'obstacle_tokens'] = 'НЕТ ИНФОРМАЦИИ'
        info.loc[i, 'sex_tokens'] = 'НЕТ ИНФОРМАЦИИ'
        info.loc[i, 'education_tokens'] = 'НЕТ ИНФОРМАЦИИ'
        info.loc[i, 'activity_tokens'] = 'НЕТ ИНФОРМАЦИИ'
        info.loc[i, 'job_tokens'] = 'НЕТ ИНФОРМАЦИИ'
```

```python
labels = pd.DataFrame(index=goals_ids)

labels['sp_0'] = [dataset.loc[dataset['label_goal_id'] == i]['label_specific'].values[0] for i in goals_ids]
labels['sp_1'] = [dataset.loc[dataset['label_goal_id'] == i]['label_specific'].values[1] for i in goals_ids]
labels['sp_2'] = [dataset.loc[dataset['label_goal_id'] == i]['label_specific'].values[2] for i in goals_ids]

labels['at_0'] = [dataset.loc[dataset['label_goal_id'] == i]['label_attainable'].values[0] for i in goals_ids]
labels['at_1'] = [dataset.loc[dataset['label_goal_id'] == i]['label_attainable'].values[1] for i in goals_ids]
labels['at_2'] = [dataset.loc[dataset['label_goal_id'] == i]['label_attainable'].values[2] for i in goals_ids]

labels['tb_0'] = [dataset.loc[dataset['label_goal_id'] == i]['label_time_bound'].values[0] for i in goals_ids]
labels['tb_1'] = [dataset.loc[dataset['label_goal_id'] == i]['label_time_bound'].values[1] for i in goals_ids]
labels['tb_2'] = [dataset.loc[dataset['label_goal_id'] == i]['label_time_bound'].values[2] for i in goals_ids]
```

```python
data_for_specificity_prediction = pd.DataFrame(index=goals_ids)
data_for_specificity_prediction['goal'] = [goals.loc[goals['id'] == i]['goal'].values[0] for i in goals_ids]
data_for_specificity_prediction.index.rename('id', inplace = True)

data_for_specificity_prediction = pd.concat([data_for_specificity_prediction, info, labels], axis=1)
```
